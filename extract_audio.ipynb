```python
import os
from natsort import natsorted
from tqdm import tqdm
```

```python
data_dir = '../Term_Project_Small'
tgt_dir = '../Term_Project_Full'

folders = natsorted(os.listdir(data_dir))
```

```python
for folder in folders:
    videos = natsorted(os.listdir(os.path.join(data_dir, folder)))

    for video in tqdm(videos):
        ID = video.split('.')[0]

        if os.path.exists(os.path.join(tgt_dir, ID)):
            os.makedirs(os.path.join(tgt_dir, ID, 'audio'), exist_ok=True)
            os.system('ffmpeg -y -i {} -async 1 {}'.format(os.path.join(data_dir, folder, video), os.path.join(tgt_dir, ID, 'audio', 'audio.wav')))
```
